```python
# Import dependencies
import pandas as pd
```

```python
# Read data into a Pandas DataFrame.
vine_table_df = pd.read_csv("vine_table.csv")
vine_table_df.head()
```

```python
# 1) Retrive rows where total_votes count is >= 20.
vine_table = vine_table_df[vine_table_df["total_votes"] >= 20]
vine_table.head()
```

```python
 # 2) Retrieve rows where the number of helpful_votes divided by total_votes is >= 50%. 
filtered_table = vine_table[vine_table["helpful_votes"]/vine_table["total_votes"] >= 0.5]
filtered_table.head()
```

```python
# 3) Retrive rows where reviews are written as part of the Vine program.
y_vine = filtered_table[filtered_table["vine"] == "Y"]
y_vine.head()
```

```python
# 4) Retrive rows where reviews are not part of the Vine program.
n_vine = filtered_table[filtered_table["vine"] == "N"]
n_vine.head()
```

```python
# 5a) Determine the total number of reviews (paid vs unpaid).
print(f"Total number of paid reviews: {y_vine.count()[0]}\n\
Total number of unpaid reviews: {n_vine.count()[0]}")
```

```python
# 5b) Determine the total number of 5-star reviews (paid vs unpaid).
y_five_star = y_vine[y_vine['star_rating'] == 5].count()[0]
n_five_star = n_vine[n_vine['star_rating'] == 5].count()[0]

print(f"Total number of paid 5-star reviews: {y_five_star}\n\
Total number of unpaid 5-star reviews: {n_five_star}")
```

```python
# 5c) Determine the total % of 5-star reviews (paid vs unpaid).
y_percent = y_five_star/y_vine.count()[0]*100
n_percent = n_five_star/n_vine.count()[0]*100

print(f"Total percent of paid 5-star reviews: {y_percent.round()}\n\
Total percent of unpaid 5-star reviews: {n_percent.round()}")
```
